```python
import pandas as pd
import numpy as np
```

```python
df_canada = pd.read_excel(
    'Canada.xlsx', 
    sheet_name='Canada by Citizenship', 
    skiprows=range(20), 
    skipfooter=2)
```

```python
print(df_canada.shape)
```

```python
df_canada.drop(['AREA', 'REG', 'DEV', 'Type', 'Coverage'], axis=1, inplace=True)
df_canada.head()
```

```python
df_canada.rename(columns={'OdName':'Country','AreaName':'Continent','RegName':'Region'}, inplace=True)
df_canada.head()
```

```python
#Verificar que todos los tipo de columna sean String
all(isinstance(column,str) for column in df_canada.columns)
```

```python
#Cambiar todas las columnas a String
df_canada.columns = list( map(str, df_canada.columns))
```

```python
#Verificar que todos los tipo de columna sean String
all(isinstance(column,str) for column in df_canada.columns)
```

```python
df_canada.set_index('Country', inplace=True)
df_canada.head()
```

```python
# Agregar columna con la sumatoria de las migraciones de cada país por año 
df_canada['Total'] = df_canada.sum(axis=1)
df_canada.head()
```

```python
years = list(map(str, range(1980, 2014)))
years
```

### Parte 2 Visualización de datos con matplotlib

```python
%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt

mpl.style.use('ggplot')
```

#### Gráficos de area

```python
df_canada.sort_values(['Total'], ascending=False, axis=0, inplace=True)

df_canada.head()
```

```python
df_top5 = df_canada.head()
df_top5
```

```python
# Transpuesta
df_top5 = df_top5[years].transpose()
df_top5.head()
```

```python
df_top5.index = df_top5.index.map(int)
df_top5.head()
```

```python
df_top5.plot( kind='area', stacked=False, figsize=(20,10))
plt.title('Tendencias de inmigracion de los primeros 5 paises a Canada')
plt.ylabel('Número de inmigrantes')
plt.xlabel('Años')
plt.show()
```

```python
df_top5.plot( kind='area', alpha=.25, stacked=False, figsize=(20,10))
plt.title('Tendencias de inmigracion de los primeros 5 paises a Canada')
plt.ylabel('Número de inmigrantes')
plt.xlabel('Años')
plt.show()
```

```python
ax= df_top5.plot(kind='area',alpha=0.35, figsize=(20,10))
ax.set_title('Tendencias de inmigracion de los 5 principales paises a Canada')
ax.set_ylabel('Número de inmigrantes')
ax.set_xlabel('Años')
```

### Gráficos de barras

```python
df_chile = df_canada.loc['Chile', years]
df_chile.head()
```

```python
# En los diagramas de barras el eje X se usa para etiquetar y la longitud de la barra corresponde al valor de la variable
df_chile.plot(kind='barh', figsize=(12,12))
plt.xlabel('Número de inmigrantes')
plt.ylabel('Años')
plt.title('Inmigrantes chilenos a canada desde 1980 al 2013')

for index, value in enumerate(df_chile):
    label = format(int(value), ',')
    plt.annotate(label, xy=(value, index), color='black')
```

```python
df_mexico = df_canada.loc['Mexico', years]
df_mexico.head()
```

```python
df_mexico.plot(kind='bar', figsize=(10,6))
plt.xlabel('Años')
plt.ylabel('Número de inmigrantes')
plt.title('Inmigrantes chilenos a canada desde 1980 al 2013')
```

### Gráficos circulares

```python
df_continent = df_canada.groupby('Continent', axis=0).sum()
df_continent
```

```python
df_continent['Total'].plot(
    kind='pie', 
    figsize=(5,6), 
    autopct='%1.1f%%', 
    #startangle=0, 
    shadow=True)

plt.title('Inmigración a canada por continente (1980-2013)')
```

```python
explode_list=[0.1,-2.5,0,0,0.1,0.3]

df_continent['Total'].plot(
    kind='pie', 
    figsize=(15,6), 
    autopct='%1.1f%%',
    #startangle=0, 
    #shadow=True, 
    labels=None, 
    pctdistance=1.12, 
    explode=explode_list)
    
plt.axis('equal')
plt.title('Inmigración a canada por continente (1980-2013)')
plt.legend(labels=df_continent.index, loc='upper right')
```
