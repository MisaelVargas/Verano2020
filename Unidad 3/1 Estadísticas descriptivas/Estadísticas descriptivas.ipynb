```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
datos = pd.read_csv('rpi_describe.csv')
```

```python
datos.head()
```

```python
datos.tail()
```

```python
datos.describe()
```

```python
# Añadir nueva columna rounded con los valores redondeados de weight
datos['rounded'] = datos.weight.round(2)

datos.head()
```

```python
# Crea nueva columna diff
datos['diff'] = datos['rounded'] - 20

datos.head()
```

```python
count = datos['rounded'].count()
mean = datos['rounded'].mean()
median = datos['rounded'].median()
std = datos['rounded'].std()
rng = datos['rounded'].max() - datos['rounded'].min()
```

```python
countstring = 'La numero de observaciones es {}, en el conjunto de datos'.format(count)

meanstring = 'La media de la distribución es {:.2f}, y la mediana es {} en el conjunto de datos'.format(mean, median)

stdstring = 'La desviacion estandar es {:.2f}, para el conjunto de datos'.format(std)

maxminstring = 'El valor mínimo es {}, el valor máximo es {}'.format(datos['rounded'].min(), datos['rounded'].max())

rangestring = 'EL rango es {:.2f}'.format(rng)
```

```python
print(countstring)
print(meanstring)
print(stdstring)
print(rangestring)
print(maxminstring)
```

```python
# Create a variable called 'freq' to hold the weight values and their frequencies
freq = datos['rounded'].value_counts()

# Convert the freq object to a data frame. Use to_frame().
freq = freq.to_frame().reset_index()
```

```python
type(freq)
```

```python
freq.columns = ['value','freq']
```

```python
freq.head()
```

```python
# Set a size for the graph
plt.figure(figsize=(20,10))

# Add axis labels
plt.ylabel('Frequency')
plt.xlabel('Weight')

# Plot the graph using a round symbol "o" of size 10
plt.plot(freq.value,freq.freq, "o", markersize = 10, color = 'b')
```
